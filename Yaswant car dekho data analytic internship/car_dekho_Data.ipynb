```python
#Data analysis case study- car dekho
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("car_Dekho_Data.csv")
data.head()
```

```python
print(data.Car_Name.value_counts())
```

```python
print(data.Fuel_Type.value_counts())
```

```python
data.Fuel_Type.value_counts().plot.bar()
```

```python
data.Fuel_Type.value_counts().plot.pie()
```

```python
fig_dims = (12,20)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(y=data.Car_Name,ax=ax, data = data)
```

```python
fig_dims = (30,30)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(x=data.Car_Name,ax=ax, data = data)
```
